```python
def simulate_guard_path(map_data):
    # Parse the map and find the starting position and direction
    directions = {'^': (-1, 0), '>': (0, 1), 'v': (1, 0), '<': (0, -1)}
    right_turn = {'^': '>', '>': 'v', 'v': '<', '<': '^'}
    
    rows = map_data.strip().split('\n')
    height, width = len(rows), len(rows[0])
    
    # Find the initial position and direction
    for r in range(height):
        for c in range(width):
            if rows[r][c] in directions:
                position = (r, c)
                direction = rows[r][c]
                break
    
    visited = set()
    visited.add(position)
    
    while True:
        # Calculate the next position
        dr, dc = directions[direction]
        next_position = (position[0] + dr, position[1] + dc)
        
        # Check if the next position is out of bounds
        if not (0 <= next_position[0] < height and 0 <= next_position[1] < width):
            break
        
        # Check if there's an obstacle
        if rows[next_position[0]][next_position[1]] == '#':
            direction = right_turn[direction]  # Turn right
        else:
            # Move forward
            position = next_position
            visited.add(position)
    
    return len(visited)

# Example map input
map_input = """
....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#...
"""

# Calculate the number of positions visited
visited_count = simulate_guard_path(map_input)
print("Number of positions visited:", visited_count)
```
